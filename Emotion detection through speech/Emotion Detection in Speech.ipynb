```python
listOfDirs = ['angry', 'calm', 'disgust', 'fearful',
              'happy', 'neutral', 'sad', 'surprised']

[features, classNames, _] = aF.dirsWavFeatureExtraction(listOfDirs, mtWin, mtStep, stWin, stStep, computeBEAT=False)

if len(featuresFemale) == 0:
    print "No data found."
else:
    numOfFeatures = features[0].shape[1]
    featureNames = ["features" + str(d + 1) for d in range(numOfFeatures)]
    aT.writeTrainDataToARFF("RAVDESSAngryFemale", features, classNames, featureNames)
    
X = np.vstack((features[0],features[1],features[2],features[3],
               features[4],features[5],features[6],features[7]))

Y = np.array([])

for i in range(8):
    cls = classNames[i]
    for j in range(len(features[i])):
        Y = np.append(Y, cls)
        
        
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=48, stratify=Y)

clf = MLPClassifier(activation='tanh', alpha=0.03, batch_size='auto', beta_1=0.9,
       beta_2=0.999, early_stopping=False, epsilon=1e-08,
       hidden_layer_sizes=(64, 64, 64), learning_rate='constant',
       learning_rate_init=0.001, max_iter=100000, momentum=0.9,
       nesterovs_momentum=True, power_t=0.5, random_state=48, shuffle=True,
       solver='lbfgs', tol=0.0001, validation_fraction=0.1, verbose=False,
       warm_start=False)


clf.fit(X_train, y_train) 

y_pred = clf.predict(X_test)
# print clf.best_score_
# print clf.best_estimator_
prediction1=clf.predict(X_test)
print('Accuracy', metrics.accuracy_score(prediction1,y_test))
```
